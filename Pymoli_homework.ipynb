```python
#binning by demographic
bins = [0, 10, 15, 20, 25, 30, 35, 40, 45]
age_cat = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
```

```python
 # Slice the data and place it into bins
    
purchase_data_pd["Age Range"] = pd.cut(purchase_data_pd["Age"], bins, labels=age_cat)
purchase_data_pd.head()
```

```python
# Create a GroupBy object based upon "Age Range"
age_cat = purchase_data_pd.groupby("Age Range")


# Find how many rows fall into each bin
print(age_cat["SN"].count())
```

```python
# Top Spenders
#Run basic calculations to find the top 5 spenders
spenders = purchase_data_pd.groupby('SN')
spender_purchase_count = spenders['Purchase ID'].count()
avg_price_spender = spenders['Price'].mean()
total_price_spender = spenders['Price'].sum()
```

```python
#Create a summary data frame to hold the results
spenders_df = pd.DataFrame({
'Purchase Count': spender_purchase_count,
'Average Purchase Price': avg_price_spender,
'Total Purchase Value': total_price_spender})
spenders_df.head()
```

```python
#Sort the total purchase value column in descending order
sorted_spenders_df = spenders_df.sort_values(('Total Purchase Value'), ascending=False)
sorted_spenders_df.head()
```
